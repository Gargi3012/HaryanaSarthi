```python
import pandas as pd
import os

raw_path = "../backend/data/raw"

def audit_dataset(file_name):
    print(f"\n📁 Auditing: {file_name}")
    
    df = pd.read_excel(os.path.join(raw_path, file_name))
    
    print("Shape:", df.shape)
    print("\nColumn Names:")
    print(df.columns.tolist())
    
    print("\nMissing Values:")
    print(df.isnull().sum())
    
    print("\nDuplicate Rows:", df.duplicated().sum())
    
    print("\nData Types:")
    print(df.dtypes)
    
    return df

haryana_df = audit_dataset("HaryanaScheme.xlsx")
```
