```python
import pandas as pd
import os

BASE_DIR = os.getcwd()

raw_path = os.path.abspath(os.path.join(BASE_DIR, "..", "backend", "data", "raw"))

file_path = os.path.join(raw_path, "Haryana_Govt_Colleges_100.xlsx")

print("Trying path:", file_path)

college_df = pd.read_excel(file_path)

print(college_df.shape)
```

```python
college_df.columns = (
    college_df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

college_df.columns
```

```python
college_df = college_df.loc[:, ~college_df.columns.duplicated()]
```

```python
college_df.info()
college_df.isnull().sum()
college_df.duplicated().sum()
```

```python
college_df['min_percentage_required'] = pd.to_numeric(
    college_df['min_percentage_required'], errors='coerce'
)

college_df['entrance_exam_required'] = college_df['entrance_exam_required'].astype(str).str.strip()
college_df['has_entrance_exam'] = college_df['has_entrance_exam'].astype(str).str.strip().str.lower()
```

```python
text_cols = [
    'college_name','college_type','location','affiliated_university',
    'accreditation','courses_offered','degree_programs_offered',
    'government_funding','mode_of_study','hostel_facilities',
    'scholarships_available','region','infrastructure',
    'admission_process','placement_assistance',
    'accredited_by','eligibility_criteria'
]

for col in text_cols:
    if col in college_df.columns:
        college_df[col] = college_df[col].astype(str).str.strip().str.lower()
```

```python
num_cols = [
    'year_established',
    'tuition_fees',
    'student_capacity',
    'ranking'
]

for col in num_cols:
    if col in college_df.columns:
        college_df[col] = pd.to_numeric(college_df[col], errors='coerce')
```

```python
college_df["college_name"] = college_df["college_name"].astype(str).str.strip()
college_df["location"] = college_df["location"].astype(str).str.strip()
college_df["region"] = college_df["region"].astype(str).str.strip()
college_df["mode_of_study"] = college_df["mode_of_study"].astype(str).str.strip()
college_df["courses_offered"] = college_df["courses_offered"].astype(str).str.strip()
college_df["degree_programs_offered"] = college_df["degree_programs_offered"].astype(str).str.strip()
college_df["entrance_exam_required"] = college_df["entrance_exam_required"].astype(str).str.strip()
college_df["has_entrance_exam"] = (
    college_df["has_entrance_exam"]
    .astype(str)
    .str.strip()
    .str.lower()
)
```

```python
college_df["min_percentage_required"] = pd.to_numeric(
    college_df["min_percentage_required"], errors="coerce"
)

college_df["tuition_fees"] = pd.to_numeric(
    college_df["tuition_fees"], errors="coerce"
)
```

```python
college_df["has_entrance_exam"] = college_df["has_entrance_exam"].replace({
    "yes": "yes",
    "y": "yes",
    "true": "yes",
    "1": "yes",
    "no": "no",
    "n": "no",
    "false": "no",
    "0": "no"
})
```

```python
college_df['has_entrance_exam'] = college_df['entrance_exam_required'].notna()
```

```python
college_df["min_qualification"] = college_df["degree_programs_offered"]
```

```python
import os

cleaned_dir = os.path.abspath(os.path.join(os.getcwd(), "..", "backend", "data", "cleaned"))
os.makedirs(cleaned_dir, exist_ok=True)

output_path = os.path.join(cleaned_dir, "colleges_cleaned.csv")

college_df.to_csv(output_path, index=False)

print("Saved successfully at:", output_path)
```
