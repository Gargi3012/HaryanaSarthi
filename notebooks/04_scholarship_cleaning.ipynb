```python
import pandas as pd
import numpy as np
import os
```

```python
raw_path = "../data/raw"
```

```python
sch_df = pd.read_excel(os.path.join(raw_path, "India_Government_Scholarships_25k_Dataset.xlsx"))
```

```python
sch_df.shape
```

```python
sch_df.head()
sch_df.info()
sch_df.isnull().sum()
sch_df.duplicated().sum()
```

```python
sch_df.columns = (
    sch_df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

sch_df.columns
```

```python
sch_df = sch_df.drop_duplicates()
```

```python
num_cols = [
    'min_class','max_class','min_marks_required',
    'annual_scholarship_amount','income_limit',
    'processing_time_days','hostel_allowance',
    'book_allowance','monthly_stipend'
]

for c in num_cols:
    if c in sch_df.columns:
        sch_df[c] = pd.to_numeric(sch_df[c], errors='coerce')
```

```python
if 'application_deadline' in sch_df.columns:
    sch_df['application_deadline'] = pd.to_datetime(sch_df['application_deadline'], errors='coerce')
```

```python
text_cols = ['ministry','scheme_type','eligible_category','state','renewal_eligible']

for c in text_cols:
    if c in sch_df.columns:
        sch_df[c] = sch_df[c].astype(str).str.strip().str.lower()
```

```python
sch_df['state'] = sch_df['state'].astype(str).str.lower().str.strip()

sch_df['is_haryana_eligible'] = sch_df['state'].isin(['haryana', 'all india'])

haryana_sch_df = sch_df[sch_df['is_haryana_eligible']].copy()

haryana_sch_df.shape
```
